```python
from manim import *
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter

class BeautifulPreferentialGraph(Scene):
    def construct(self):
        initial_nodes = 5
        total_nodes = 100  # Moderate size for visualization
        m_edges = 3

        # Create the graph using the Barabsi-Albert model
        G = nx.barabasi_albert_graph(total_nodes, m_edges)
        pos = nx.spring_layout(G, seed=42)  # Spring layout for better spacing

        node_dots = {}
        edge_lines = []

        # Scaling factor to fit the graph nicely on the screen
        scale_factor = 4  
        normal_node_size = 0.08  # Standard node size (slightly bigger)
        hub_node_size = 0.20  # Bigger size for central nodes (even larger now)

        self.wait(1)  # Small delay before animation starts

        # Animate nodes appearing one by one
        max_degree = max(dict(G.degree()).values())  # Find max degree for scaling
        for node in G.nodes():
            x, y = pos[node]
            scaled_position = [x * scale_factor, y * scale_factor, 0]

            # Determine if node is a hub (has high degree)
            degree = G.degree(node)
            node_size = hub_node_size if degree > max_degree * 0.3 else normal_node_size  # Bigger size for hubs
            color = interpolate_color(BLUE, PINK, degree / max_degree)

            dot = Dot(point=scaled_position, radius=node_size, color=color)
            node_dots[node] = dot

            # Animate node appearance with a soft glow effect
            self.play(GrowFromCenter(dot), run_time=0.5)

            # Draw edges dynamically after the node appears
            for neighbor in G.neighbors(node):
                if neighbor in node_dots:  # Ensure the neighbor has already been added
                    edge_line = Line(node_dots[node].get_center(), node_dots[neighbor].get_center(), color=WHITE, stroke_width=1.2)
                    edge_lines.append(edge_line)
                    self.play(Create(edge_line), run_time=0.2, rate_func=smooth)

        self.wait(2)

        # Plot degree distribution after visualization
        self.plot_degree_distribution(G)

    def plot_degree_distribution(self, G):
        degrees = [G.degree(n) for n in G.nodes()]
        degree_count = Counter(degrees)
        deg, cnt = zip(*sorted(degree_count.items()))

        plt.figure(figsize=(6, 4))
        plt.plot(deg, cnt, 'bo-', markersize=5)
        plt.xscale("log")
        plt.yscale("log")
        plt.xlabel("Degree (log scale)")
        plt.ylabel("Frequency (log scale)")
        plt.title("Power-Law Degree Distribution")
        plt.grid(True, linestyle="--", linewidth=0.5)
        plt.show()

# Run inside Jupyter Notebook
%manim -ql BeautifulPreferentialGraph
```
